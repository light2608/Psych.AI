# 1. THIS DETECTS FROM THE SAMPLE VIDEO

```python
import cv2
import mediapipe as mp
import numpy as np

# MediaPipe modules
mp_drawing = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles
mp_holistic = mp.solutions.holistic

video_path1 = 'C:/Users/light/OneDrive/Desktop/PsychAI/Final_output video/Lucy/Lucy_depression full video.mp4'
video_path2 = 'C:/Users/light/OneDrive/Desktop/PsychAI/Final_output video/Lucy/Lucy_depression.mp4'

cap1 = cv2.VideoCapture(video_path1)
cap2 = cv2.VideoCapture(video_path2)

# Ensure same frame rate (choose one of the videos' frame rate)
fps = cap1.get(cv2.CAP_PROP_FPS)
print(f"FPS: {fps}")


# Output dimensions
output_width = 1700
output_height = 550

# Initialize MediaPipe Holistic
holistic = mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5)

def process_frame(frame, holistic):
    image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    image.flags.writeable = False

    results = holistic.process(image)

    image.flags.writeable = True
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

    if results.face_landmarks:
        mp_drawing.draw_landmarks(
            image, results.face_landmarks, mp_holistic.FACEMESH_CONTOURS,
            landmark_drawing_spec=None,
            connection_drawing_spec=mp_drawing_styles.get_default_face_mesh_contours_style())
    if results.pose_landmarks:
        mp_drawing.draw_landmarks(
            image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS,
            landmark_drawing_spec=mp_drawing_styles.get_default_pose_landmarks_style())
    if results.left_hand_landmarks:
        mp_drawing.draw_landmarks(
            image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS,
            landmark_drawing_spec=mp_drawing_styles.get_default_hand_landmarks_style(),
            connection_drawing_spec=mp_drawing_styles.get_default_hand_connections_style())
    if results.right_hand_landmarks:
        mp_drawing.draw_landmarks(
            image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS,
            landmark_drawing_spec=mp_drawing_styles.get_default_hand_landmarks_style(),
            connection_drawing_spec=mp_drawing_styles.get_default_hand_connections_style())
    return image



while cap1.isOpened() and cap2.isOpened():
    ret1, frame1 = cap1.read()
    ret2, frame2 = cap2.read()

    # Break if either video ends or fails to read a frame
    if not ret1:
        print(f"Error reading frame from video 1 at frame number: {cap1.get(cv2.CAP_PROP_POS_FRAMES)}")
        break
    if not ret2:
        print(f"Error reading frame from video 2 at frame number: {cap2.get(cv2.CAP_PROP_POS_FRAMES)}")
        break  

    # Resize frames before processing
    frame1 = cv2.resize(frame1, (output_width // 2, output_height))  
    frame2 = cv2.resize(frame2, (output_width // 2, output_height))
    
    # Process frame2 (with landmarks)
    image2 = process_frame(frame2, holistic)
   

    # Combine the two frames horizontally
    combined_frame = np.hstack((frame1, image2))

    # Show the combined frame
    cv2.imshow('Combined Videos', combined_frame)

    if cv2.waitKey(25) & 0xFF == ord('q'):  # Adjust the waitKey value as needed
        break

cap1.release()
cap2.release()
cv2.destroyAllWindows()
```

# 1. THIS IS DETECTING REAL TIME

```python
import cv2
import mediapipe as mp

# MediaPipe drawing and holistic modules
mp_drawing = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles
mp_holistic = mp.solutions.holistic

# Accessing the webcam (0 usually represents the default camera)
cap = cv2.VideoCapture(0)  

# Initialize holistic model for combined pose, face, and hand tracking
with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:

    while cap.isOpened():
        ret, frame = cap.read()

        # Check if a frame was successfully captured
        if not ret:
            break

        # Recolor Feed to RGB for MediaPipe
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False

        # Process the image to detect landmarks
        results = holistic.process(image)

        # Recolor back to BGR for OpenCV rendering
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        # 1. Draw Face Landmarks
        mp_drawing.draw_landmarks(
            image,
            results.face_landmarks,
            mp_holistic.FACEMESH_CONTOURS,  # Correct connection reference
            mp_drawing.DrawingSpec(color=(80, 110, 10), thickness=1, circle_radius=1),
            mp_drawing.DrawingSpec(color=(80, 256, 121), thickness=1, circle_radius=1),
        )

        # 2. Right Hand Landmarks
        mp_drawing.draw_landmarks(
            image,
            results.right_hand_landmarks,
            mp_holistic.HAND_CONNECTIONS,
            mp_drawing_styles.get_default_hand_landmarks_style(),  
            mp_drawing_styles.get_default_hand_connections_style() 
        )

        # 3. Left Hand Landmarks
        mp_drawing.draw_landmarks(
            image,
            results.left_hand_landmarks,
            mp_holistic.HAND_CONNECTIONS,
            mp_drawing_styles.get_default_hand_landmarks_style(), 
            mp_drawing_styles.get_default_hand_connections_style() 
        )

        # 4. Pose Landmarks
        mp_drawing.draw_landmarks(
            image,
            results.pose_landmarks,
            mp_holistic.POSE_CONNECTIONS,
            landmark_drawing_spec=mp_drawing_styles.get_default_pose_landmarks_style(),
            connection_drawing_spec=mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2)  # Customized for connections
)

    
        # Display the annotated image
        cv2.imshow('Raw Webcam Feed', image)

        # Break the loop when 'q' is pressed
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

# Release resources
cap.release()
cv2.destroyAllWindows()
```

## 3.2 Train Machine Learning Classification Model

```python
from sklearn.pipeline import make_pipeline 
from sklearn.preprocessing import StandardScaler 

from sklearn.linear_model import LogisticRegression, RidgeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
```

```python
pipelines = {
    'lr':make_pipeline(StandardScaler(), LogisticRegression()),
    'rc':make_pipeline(StandardScaler(), RidgeClassifier()),
    'rf':make_pipeline(StandardScaler(), RandomForestClassifier()),
    'gb':make_pipeline(StandardScaler(), GradientBoostingClassifier()),
}
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
fit_models = {}
for algo, pipeline in pipelines.items():
    model = pipeline.fit(X_train, y_train)
    fit_models[algo] = model
```

```python
fit_models['rc'].predict(X_test)
```

## 3.3 Evaluate and Serialize Model
